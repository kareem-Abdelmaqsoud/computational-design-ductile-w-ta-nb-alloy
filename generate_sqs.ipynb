```python
import numpy as np
from tqdm import tqdm
from ase.build import bulk
from icet import ClusterSpace
from icet.tools.structure_generation import generate_sqs
# Base structure
primitive_structure = bulk('W', cubic=True)

# Define cluster space
cs = ClusterSpace(primitive_structure, [8.0, 4.0], ['W', 'Ta', 'Nb'])

sqs_structures = {}

# Enumerate all integer combinations of W, Ta, Nb that sum to 10
total_atoms = 10
compositions = []
for n_W in range(total_atoms + 1):
    for n_Ta in range(total_atoms + 1 - n_W):
        n_Nb = total_atoms - n_W - n_Ta
        compositions.append((n_W, n_Ta, n_Nb))

# Generate SQS for each composition
for (n_W, n_Ta, n_Nb) in tqdm(compositions):
    # Skip trivial single-element cases if desired
    if len([x for x in [n_W, n_Ta, n_Nb] if x > 0]) == 1:
        continue

    target_concentrations = {
        'W': n_W / total_atoms,
        'Ta': n_Ta / total_atoms,
        'Nb': n_Nb / total_atoms
    }

    print("target_concentrations:", target_concentrations)
    try:
        sqs = generate_sqs(
            cluster_space=cs,
            max_size=total_atoms,
            n_steps=10000,
            target_concentrations=target_concentrations
        )
        sqs_structures[(n_W, n_Ta, n_Nb)] = sqs
    except Exception as e:
        print(f"Failed for composition (W={n_W}, Ta={n_Ta}, Nb={n_Nb}): {e}")
```

```python
import json
from pymatgen.io.ase import AseAtomsAdaptor

# Convert tuple keys to strings and structures to dicts
pmg_structures = {str(frac): AseAtomsAdaptor.get_structure(sqs).as_dict()
                  for frac, sqs in sqs_structures.items()}

# Save to JSON
with open("sqs_structures_full_triangle.json", "w") as f:
    json.dump(pmg_structures, f, indent=2)
```
