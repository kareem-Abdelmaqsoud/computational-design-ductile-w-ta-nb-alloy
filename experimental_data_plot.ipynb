```python
import matplotlib.pyplot as plt
import numpy as np

## results for W-Ta-Nb with 10% Nb
## DFT results
dft_tungsten_fractions =  [0. , 10, 20, 30, 50, 60, 70, 80, 100]
dft_pugh_ratios = [0.31093764, 0.30292644, 0.31076683, 0.29909123, 0.32965786,
       0.42543827, 0.45838539, 0.46637931, 0.51238685]
## Experimental results
experimental_tungsten_fractions = [30, 43, 61]
experimental_crack_fractions = [0, 2, 12]

# Create figure and first y-axis
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot Pugh ratio on left y-axis
color1 = 'tab:blue'
ax1.set_xlabel('Tungsten (Atomic %)', fontsize=16)
ax1.set_ylabel('Pugh Ratio (G/B)', color=color1, fontsize=16)
line1 = ax1.plot(dft_tungsten_fractions, dft_pugh_ratios, 'o-', color=color1, 
                 markersize=8, linewidth=2, label='DFT Pugh Ratio')
ax1.tick_params(axis='y', labelcolor=color1, labelsize=14)
ax1.tick_params(axis='x', labelsize=14)
ax1.grid(alpha=0.3)

# Create second y-axis
ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Crack Fraction (mm/mm)', color=color2, fontsize=16)

# Plot experimental points with different markers
for i, (frac, crack) in enumerate(zip(experimental_tungsten_fractions, experimental_crack_fractions)):
    if frac == 43:
        ax2.scatter(frac, crack, color=color2, s=300, marker='*', label='Experimental Crack Fraction')
    else:
        ax2.scatter(frac, crack, color=color2, s=100, marker='s')

ax2.tick_params(axis='y', labelcolor=color2, labelsize=14)
ax2.set_ylim(-1, 20)
ax2.set_yticks(np.arange(0, 21, 2))

# Combine legends in one box
lines1, labels1 = ax1.get_legend_handles_labels()
# Get only one label for experimental data
legend_elements = lines1 + [plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color2, 
                                      markersize=8, label='Experimental Crack Fraction')]
ax1.legend(handles=legend_elements, loc='upper left', fontsize=12)

# plt.title('Pugh Ratio vs Crack Fraction for W-Ta-Nb with 10% Nb', fontsize=16)
plt.tight_layout()
plt.show()
```
